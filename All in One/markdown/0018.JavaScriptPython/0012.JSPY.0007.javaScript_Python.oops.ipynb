# class
Classes are a template for creating objects. They encapsulate data with code to work on that data. Classes in JS are built on prototypes but also have some syntax and semantics that are unique to classes.

#### syntax :- javascript

> with global variable

```js
class ClassName {
  arg0; // global variable
  arg1; // global variable
  ....
  argN; // global variable
  // if you create constructor , then you may omit to declare global variable 
  constructor(arg0, arg1, .... , argN) { 
    this.arg0 = arg0;
    this.arg0 /*this.arg0 :- global variable*/= arg0; /*arg0 :- argument by constructorglobal variable*/
    this.arg0 = arg0;
          .....
    this.arg0 = arg0;
   }
  method_1() { ... }
  method_2() { ... }
  method_3() { ... }
}
```

> without global variable

```js
class ClassName {
  constructor(arg0, arg1, .... , argN) { 
    this.arg0 = arg0;
    this.arg0 /*this.arg0 :- global variable*/= arg0; /*arg0 :- argument by constructorglobal variable*/
    this.arg0 = arg0;
          .....
    this.arg0 = arg0;
   }
  method_1() { ... }
  method_2() { ... }
  method_3() { ... }
}
```

> **function**
* don't write `function method_1(){....}` , 
  * because inside class we can't use 'function' keyword.
  * if we create function/method inside class then we go diretly like
    * ` methodName() { ... }  `
  * inside function if you are creating 
    * any variable then use :- `let/var/const variableName = value;`
      * or go with normal variable declaration method.
    * any function then use :- ` function methodName() { ... }` 
      * or go with normal function declaration method.
> **constructor**
* inside constructor we can see 
* this.arg0 = arg0;
  * 
    ```js
    this.arg0 /*this.arg0 :- global variable*/= arg0; /*arg0 :- argument by constructorglobal variable*/
    ```
> **class**
* Class methods are created with the same syntax as object methods.
* Use the keyword class to create a class.
* Always add a constructor() method.
