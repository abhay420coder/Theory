* install react globally in your pc
  * `npm install -global create-react-app` 
 
* create react app
  * `npx create-react-app apppName`


* Starts the development server.
  * `npm start`  :-  

* Bundles the app into static files for production.
  * `npm run build`
    
* Starts the test runner.
* `npm test`
