* if an expression contains mixed variable types ,
  * then the return value is of the most complex type contained within the expression.

#### Operator Overloading

* The term **operator overloading** refers to the ability of a language
  * to perform different operations on different object types
  * when using the same operator.
* From a programming prespective ,
  * the ability of a language to support operator overloading allows for a simplified programming process.

> * For example , the + operator is used to perform below operation
>   * add two numbers
>   * concat two string

> * For example , the [:] operator is used to perform below operation
>   * extract slices from strings
>   * extract slices from tuples
>   * extract slices from lists

* The operator performs the same operation on each object ,
  * even through the objects store and use their informationin different ways.

  * if the argument `z` is add
  * then the return-value is  `pow(x, y)%z`

  * Round the floating-point number `x` to `0` digits

    * Use `int` to convert a floating-point number to an integer ,

### Formatting (Format in String)

* The `%` operator in Python allows you to format strings and other objects using the same basic notation as supported by the **sprintf** function in C.
*
```py
    # general method
    print('....... formatCode1 ..... formatCode2 ....  ...... formatCodeN.... ' , %(value1 , value2 , ..... , valueN))

    # dictionary method
    dictVariable = {"key1":value1, "key2":value2, ...... , "keyN":valueN }
    print('....... %(key1)formatCode1 ..... %(key2)formatCode2 ....  ...... %(keyN)formatCode1 .... ' , %(value1 , value2 , ..... , valueN))

    # format method
    print('....... {} ..... {} ....  ...... Ntimes curlyBracket {}.... '.format(value1 , value2 , ..... , valueN))

```

#### String Concatenation

* you can not  concatenate string with other objects.

* Method-1
  * Putting More than one quoted constant on a line results in the constants being. concatenated.
* Method-2
  * Concatenate string objects and/or constants using the `+` operator.
* Method-3
  * String can also be multiplied(repeated) using `*` operator.



````python
# Method-1 :- Putting More than one quoted constant on a line results in the constants being.
string1 = 'i' 'am' 'the' 'warriors'
string2 = 'i' ' am' ' the' ' warriors'
print("string1    :-  " , string1) # iamthewarriors
print("string2    :-  " , string2) # i am the warriors

# Method-2 :- Concatenate string objects and/or constants using the `+` operator.
string3 = 'i'+'am'+'the'+'warriors'
string4 = 'i'+' am'+' the'+' warriors'
print("string3    :-  " , string3) # iamthewarriors
print("string4    :-  " , string4) # i am the warriors

# you can not  concatenate string with other objects.
# string5 = 'i' ' am' ' the' ' warriors' 45
# print("string5    :-  " , string5) # error
# string6 = 'i'+' am'+' the'+' warriors' + 45
# print("string6    :-  " , string6) # error

# Method-3 :- String can also be multiplied(repeated) using `*` operator.
string7 = 'Abhay '
print("string7    :-  " , string7*5) # Abhay Abhay Abhay Abhay Abhay 
````

#### String slicing
* `string[start : end]` 
  * contains all the characters starting from **start Index**  up until but not including **end Index**.
* `string[ : end]` 
  * contains all the characters starting from **0th Index** up until but not including end.
* `string[start : ]` 
  * contains all the characters starting from **start**  up until **last Index**.

````python
string8 = 'I returned a bag of groceries'
print("string8[2:10]    :-  " , string8[2:10])  # returned
print("string8[13:]    :-  " , string8[13:]) # bag of groceries

print('index [-5 : ] of string8 :-    ' , string8[-5 : ]) # eries
print('index [ : -5] of string8 :-    ' , string8[ : -5]) # I returned a bag of groc

````
