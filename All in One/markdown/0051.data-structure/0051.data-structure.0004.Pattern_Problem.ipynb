````python
# with for loop
def printPyramid(row):
  
  # count `star` and `space` with respect to `n` only for row 1 (`first row`) .
  nst = 1 
  nsp = row - 1
  
  for i in range(row):
    
    # write loop to print `space`
    for j in range(nsp):
      print(" " , end='')
    # write loop to print `star`
    for j in range(nst):
      if(j!=(nst-1)):
        print("*" , end=' ')
      else:
        print("*" , end='')
    print()
    
    # relation in between `row 1` and `row 2` for `nst` (number of star) and `nsp` (number of space) for doing this use table.
    nst = nst + 1
    nsp = nsp - 1

printPyramid(5)
    
````

````python
# with recursion
# in recursion :- one for loop = one recursion function

  
  
def printPyramid(row , n = 0):
  # base condition
  if n==row :
    return
  else:
    # recursion statement
    nst = n+1
    nsp = row-(n+1)
    printSpace(nsp)
    printStar(nst)
    
    print()
    printPyramid(row , n+1)
    
def printStar(nst , n=0):
  # base condition
  if(n == nst):
    return
  else:
    # recursion statement
    if(n == (nst-1)):
      print("*" , end='')
    else :
      print("*" , end=' ')
    printStar(nst , n+1)
    
def printSpace(nsp , n=0):
  # base condition
  if(n == nsp):
    return
  else:
    # recursion statement
    print(" " , end='')
    printSpace(nsp , n+1)  
    
     
printPyramid(5)
    
````

````python
def printDiamondShapePattorn(row):
  nst = 1; 
  nsp = row-1
  for i in range(((2*row)-1)):
    for j in range(nsp):
      print(" " , end='')
    for k in range(nst):
      if (k!=nst):
        print("*" , end='')
    print()
    if(i<(row-1)):
      nst = nst+2
      nsp = nsp-1
    else:
      nst = nst-2
      nsp = nsp+1

printDiamondShapePattorn(5)
````

````python
def printMirroredRightTriangle(row):
  nst = 1
  nsp = row-1
  for i in range(row):
    for j in range(nsp):
      print(" " , end='')
    for j in range(nst):
      if (j!=(nst-1)):
        print('*' , end='')
      else :
        print('*' , end='')
    print()
    nst = nst+1
    nsp = nsp-1
printMirroredRightTriangle(6)
````
